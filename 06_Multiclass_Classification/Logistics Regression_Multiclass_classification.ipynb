```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### Load Data

```python
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = iris.frame
df.head()
```

```python
df["target_names"] = df["target"].map({0:"setosa", 1:"versicolor", 2:"virginica"})
df.head()
```

```python
df.columns
```

```python
iris.target.unique()
```

### Data Visualization and Exploration

```python
sns.scatterplot(df,x = "sepal length (cm)" , y ="sepal width (cm)" , hue = "target_names" )
```

```python
sns.scatterplot(df,x = "petal length (cm)" , y ="sepal width (cm)" , hue = "target_names")
```

```python
X = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']]
y = df["target"]
X.head()
```

### Model Training (Train_test_Split)

```python
from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 5)
```

### import Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
```

### Model Evaluation Accuracy_score

```python
model.score(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import accuracy_score

accuracy_score( y_test, y_predicted)
```

### Model Evaluation precision, recall , F-1 Score

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_predicted)
print(report)
```

### Model Evaluation Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicted)
```

```python
sns.heatmap(cm , xticklabels = iris.target_names, yticklabels = iris.target_names, annot=True)

plt.xlabel("Predicted Labels") 
plt.ylabel("True labels")
plt.title("Confusion matrix")
```
